```python
# going to go ahead and scale data now to aid in final visuals and prepare for PCA and KMeans later
st_scale = StandardScaler()
Z = st_scale.fit_transform(data_clean_nozero)
# Z = st_scale.fit_transform(df_clean_new)
```

```python
# loop through a rnage of k's and plot the calcalted distortions of each k

dist = []

for k in range(1,12):
    kme = KMeans(n_clusters = k, n_init = 11, random_state = 42)
    
    kme.fit(Z)
    dist.append(kme.inertia_)
```

```python
plt.plot(range(1,12), dist, marker = 'o', markersize = 10, linewidth = .75)
plt.xlabel("Cluster Num.")
plt.ylabel("Dist.")
plt.tight_layout()
plt.show()
```

##### optimal number of clusters is 3 based on elbow plot methodology

```python
kmn = KMeans(n_clusters = 6, random_state = 42, algorithm='elkan')
km_clusters = kmn.fit_predict(Z)+1
df_clean_new['km_cluster'] = km_clusters
data_clean_nozero['km_cluster'] = km_clusters
```

```python
# Informs us that for the first 3 PC's and account for 91-92% of the virnce in the dataset
for i in range (1,4):
    
    pca = PCA(n_components=i)
    pca.fit(data_clean_nozero)
    print ("PC number is"+" "+ str(i) +" "+ str(pca.explained_variance_ratio_))
```

```python
def sil(clusters,ax):
    cluster_labels = np.unique(clusters)
    n_clusters = cluster_labels.shape[0]
    sil_vals = silhouette_samples(Z, clusters, metric = 'euclidean')
    y_ax_lower, y_ax_upper = 0,0
    y_ticks = []
    
    for i, c in enumerate(cluster_labels):
        c_sil_vals = sil_vals[clusters == c]
        c_sil_vals.sort()
        y_ax_upper += len(c_sil_vals)
        color = cm.viridis(float(i)/ n_clusters)
        ax.barh(range(y_ax_lower, y_ax_upper), c_sil_vals, height = 1.0, 
                edgecolor = 'none', color = color)
        y_ticks.append((y_ax_lower+y_ax_upper)/2.0)
        y_ax_lower += len(c_sil_vals)
    sil_avg = np.mean(sil_vals)
    ax.axvline(sil_avg, color = 'red', linestyle = '--', linewidth=1.5)
    
    ax.set_yticks(y_ticks)
    ax.set_yticklabels(cluster_labels +1)
    ax.set_ylabel('Cluster')
    ax.set_xlabel('Silhouette Coefficient')
```

```python
fig, (ax1)=plt.subplots(1,figsize=(8,6))

sil(km_clusters, ax1);
ax1.set_title('K-Means with Three CLusters');
plt.tight_layout()
```
